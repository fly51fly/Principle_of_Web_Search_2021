```DOM```：```HTML```和```XML```文档的编程接口，提供了对文档的结构化的表述，并定义了一种方式可以使从程序中对该结构进行访问，从而改变文档的结构，样式和内容。

### 3、用```Python```编程实现简单关键词检索

````python
with open('titles.txt', 'r') as f:
    title_list = f.read().split('\n')
````

#### 用```in```过滤包含关键词的标题 

````python
keyword = '手机'
result_list = []
for tid, title in enumerate(title_list):
    if keyword.lower() in title.lower():
        result_list.append(tid)
````

````python
count = 1
for item in result_list:
    print(f"{count}、{title_list[item]}")
    count += 1
````

#### 增加高亮函数，用```IPython.core.display```里的```HTML```实现高亮显示：

````python
from IPython.core.display import display, HTML

def highlight(text, keyword):
    idx = text.lower().find(keyword.lower())
    result = text
    if idx >= 0:
        ori_keyword = text[idx:idx+len(keyword)]
        result = text.replace(ori_keyword, f'<span style="color:red">{ori_keyword}</span>')
    return result

count = 1
for item in result_list:
    display(HTML(f"{count}、{highlight(title_list[item], keyword)}"))
    count += 1
````

#### 增加打分(关键词命中次数统计)函数，并按打分排序：

````python
def score(text, keyword):
    result = text.lower().count(keyword.lower())
    return result

keyword = '手机'
result_list = []
for tid, title in enumerate(title_list):
    if keyword.lower() in title.lower():
        result_list.append([tid, score(title, keyword)])
result_list.sort(key = lambda x: x[1], reverse=True)
````

````python
count = 1
for item in result_list:
    display(HTML(f"{count}、{highlight(title_list[item[0]], keyword)}"))
    count += 1
````

#### 更进一步封装成类，目的仍然是三点：*封装、重用、可读性*

````python
from IPython.core.display import display, HTML

class MySearchV0():
    def __init__(self, filename):
        self.docs = []
        self.load_data(filename)
        
    def load_data(self, filename):
        with open(filename, 'r') as f:
            self.docs = f.read().split('\n')
    
    def highlight(self, text, keyword):
        idx = text.lower().find(keyword.lower())
        result = text
        if idx >= 0:
            ori_keyword = text[idx:idx+len(keyword)]
            result = text.replace(ori_keyword, f'<span style="color:red">{ori_keyword}</span>')
        return result

    def score(self, text, keyword):
        result = text.lower().count(keyword.lower())
        return result

    def search(self, keyword):
        result_list = []
        for tid, title in enumerate(self.docs):
            if keyword.lower() in title.lower():
                result_list.append([tid, self.score(title, keyword)])
            result_list.sort(key = lambda x: x[1], reverse=True)
        return [doc_id for doc_id, _ in result_list]

    def render(self, result_list, keyword):
        count = 1
        for item in result_list:
            display(HTML(f"{count}、{self.highlight(self.docs[item], keyword)}"))
            count += 1
````

````python
searcher = MySearchV0('titles.txt')
````

````python
keyword = '手机'
search_result = searcher.search(keyword)
searcher.render(search_result, keyword)
````

#### 目前简单关键词搜索功能上正常，还需要从性能上进行测试  
#### 为搜索类增加倍增数据的功能，并对倍增后的搜索性能进行测试

````python
from IPython.core.display import display, HTML

class MySearchV1():
    def __init__(self, filename, scale_factor=1):
        self.docs = []
        self.scale_factor = scale_factor #数据倍增系数，默认为1
        self.load_data(filename)
        
    def load_data(self, filename):
        with open(filename, 'r') as f:
            self.docs = f.read().split('\n')
        self.docs = self.docs * self.scale_factor #简单的数据倍增
    
    def highlight(self, text, keyword):
        idx = text.lower().find(keyword.lower())
        result = text
        if idx >= 0:
            ori_keyword = text[idx:idx+len(keyword)]
            result = text.replace(ori_keyword, f'<span style="color:red">{ori_keyword}</span>')
        return result

    def score(self, text, keyword):
        result = text.lower().count(keyword.lower())
        return result

    def search(self, keyword):
        result_list = []
        for tid, title in enumerate(self.docs):
            if keyword.lower() in title.lower():
                result_list.append([tid, self.score(title, keyword)])
            result_list.sort(key = lambda x: x[1], reverse=True)
        return [doc_id for doc_id, _ in result_list]

    def render(self, result_list, keyword):
        count = 1
        for item in result_list:
            display(HTML(f"{count}、{self.highlight(self.docs[item], keyword)}"))
            count += 1
````

#### 对搜索用时进行测试
